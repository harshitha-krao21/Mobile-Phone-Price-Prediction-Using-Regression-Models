```python
#Removing text from all the below patterns from Android_version column and converting to float
patterns = ["9.0 (Pie)","8.1 (Oreo)","6.0 (Marshmallow)","7.1.1 (Nougat)","3.0 (Honeycomb)","5.1 (Lollipop)","9 (Pie)","8.0 (Oreo)"]
df['Android_version'] = df["Android_version"].astype(str)

def convert_version(version):
    # Extract the part before any space or parenthesis
    numeric_part = version.split()[0]
    parts = numeric_part.split('.')
    if len(parts) > 1:
        new_version = '.'.join(parts[:2])
    else:
        new_version = numeric_part
    return float(new_version)

# Apply the function to the 'Android_version' column
df['Android_version'] = df['Android_version'].apply(convert_version)

df['Android_version'].value_counts()
df['Android_version'].count()
```

```python
def extract_numeric(memory):
    if isinstance(memory, str):
        match = re.search(r'\d+', memory)
        if match:
            numeric_part = int(match.group(0))
            if 'TB' in memory:
                numeric_part *= 1024  # Convert TB to GB
            return int(numeric_part)
    elif isinstance(memory, int) and not np.isnan(memory):
        return int(memory)  # Convert int directly to int if not NaN
    return np.nan  # Return NaN if memory is NaN or not convertible

# Apply the function to the 'Inbuilt_memory' column
df['Inbuilt_memory'] = df['Inbuilt_memory'].apply(extract_numeric)
```

```python
#Rearranging target column "Price" in encoded dataframe
pr = df['Price']
df.drop(['Price'], axis=1, inplace = True)
df['Price'] = pr
df.head(5)
```

```python
#Seperating the dataframe into training and testing set
X_lr = df.iloc[:, :-1].values
y_lr = df.iloc[:, -1].values

#Splitting the dataset into the Training set and Test set
X_train_lr, X_test_lr, y_train_lr, y_test_lr = train_test_split(X_lr, y_lr, test_size = 0.2, random_state = 42)
```

```python
#a)Train at least two Linear Regression model with different regularization methods predicting the price using the data set.
#b)Describe the model selection method you used in traning process, and report your best models (i.e. report sklearn.model_selection.GridSearchCV.cv_results_)

#Define the parameter grid for both Lasso and Ridge
param_grid = {
    'alpha': [0.01, 0.1, 1, 10, 100]
}
```

```python
#Lasso
lasso = Lasso(max_iter=5000)
lasso_grid_search = GridSearchCV(lasso, param_grid, cv=5)
lasso_grid_search.fit(X_train_lr,y_train_lr)

#Print the best parameters and best scores
print("Lasso Best Parameters:", lasso_grid_search.best_params_)
print("Lasso Best Score (negative MSE):", lasso_grid_search.best_score_)
print("Lasso Best Estimator:", lasso_grid_search.best_estimator_)

best_lasso_model = lasso_grid_search.best_estimator_
```

```python
#Ridge
ridge = Ridge(max_iter=5000)
ridge_grid_search = GridSearchCV(ridge, param_grid, cv=5)
ridge_grid_search.fit(X_train_lr,y_train_lr)

#Print the best parameters and best scores
print("Ridge Best Parameters:", ridge_grid_search.best_params_)
print("Ridge Best Score (negative MSE):", ridge_grid_search.best_score_)
print("Ridge Best Estimator:", ridge_grid_search.best_estimator_)

best_ridge_model = ridge_grid_search.best_estimator_
```

```python
#Print the cross-validation results
print("\nLasso GridSearchCV Results:")
lasso_results_df = pd.DataFrame(lasso_grid_search.cv_results_)
print(lasso_results_df)

print("\nRidge GridSearchCV Results:")
ridge_results_df = pd.DataFrame(ridge_grid_search.cv_results_)
print(ridge_results_df)
```

```python
def extract_numeric(memory):
    if isinstance(memory, str):
        match = re.search(r'\d+', memory)
        if match:
            numeric_part = int(match.group(0))
            if 'TB' in memory:
                numeric_part *= 1024  # Convert TB to GB
            return int(numeric_part)
    elif isinstance(memory, int) and not np.isnan(memory):
        return int(memory)  # Convert int directly to int if not NaN
    return np.nan  # Return NaN if memory is NaN or not convertible

# Apply the function to the 'Inbuilt_memory' column
df_new['Inbuilt_memory'] = df_new['Inbuilt_memory'].apply(extract_numeric)
```
